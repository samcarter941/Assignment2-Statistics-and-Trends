```python
# first of all we load our dataset
import pandas as pd

# first we have to load tha dataset using read_csv
data = pd.read_csv("aapl_2014_2023.csv")

# Display the first few rows of the dataset
print(data)
print(data.head())
```

```python
# Piechart
import matplotlib.pyplot as plt

# Count the occurrences of each trend
trend_counts = data['target'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(trend_counts, labels=trend_counts.index, autopct='%1.1f%%', colors=['green', 'red', 'gray'])
plt.title('Distribution of Market Trends')
plt.show()
```

```python
# scatter graph
plt.scatter(data['open'], data['close'], alpha=0.5, color='blue')
plt.title('Scatter Plot of Open and Close Prices')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.show()
```

```python
# Box plot for every numeric feature regarding target column

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the apple dataset
df = pd.read_csv("aapl_2014_2023.csv")


# Drop the date column as it won't be used in the plot
df = df.drop(columns=['date'])

# Determine the layout based on the number of numeric features
num_features = len(df.columns) - 1  # Exclude the 'target' column
num_rows = (num_features - 1) // 4 + 1
num_cols = min(num_features, 4)

# Create box plots for each numeric feature with respect to the 'target' column
plt.figure(figsize=(18, 4 * num_rows))
for i, column in enumerate(df.columns[:-1]):  # Exclude the 'target' column
    plt.subplot(num_rows, num_cols, i + 1)
    sns.boxplot(x='target', y=column, data=df, palette='viridis')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()

```
